```python
from transformers import pipeline
```

```python
generator = pipeline('text-generation',model='EleutherAI/gpt-neo-2.7B')
```

```python
def main():
    print("Welcome to the interactive Story Generator!")

    user_name = input("Enter your username: ")
    print(f"Hello, {user_name}! Let's create a story together.")

    lang = input("Provide the genre you want to build the story on:")

    match lang:

      case "Mystery":
        
        user_choice = input("Provide a starting point or idea for the story: ")
        prompt = f'''The last echoes of her farewell party hung heavy in the air as{user_name} crossed the threshold, leaving behind the world she knew for a chase whispered only in shadows.{user_choice}'''
        res =generator(prompt,max_length=1500,do_sample=True,temperature=0.9)
        print(res[0]['generated_text'])
      case "Romantic":
        
        user_choice = input("Provide a starting point or idea for the story: ")
        prompt = f"Under the cloak of a star-dusted sky, {user_name} typed hello into the chat window, a universe of possibilities shimmering on the screen like fireflies in a moonlit meadow.{user_choice}"
        res =generator(prompt,max_length=1500,do_sample=True,temperature=0.9)
        print(res[0]['generated_text'])
      case "Horror":
        
        user_choice = input("Provide a starting point or idea for the story: ")
        prompt = f'''{user_name} found the diary hidden under the floorboards, its brittle pages filled with her own handwriting, detailing unspeakable events from a life she didn't remember living. The chilling truth dawned on her: she was haunted by herself..{user_choice}'''
        res =generator(prompt,max_length=1500,do_sample=True,temperature=0.9)
        print(res[0]['generated_text'])

      case "Sci-Fi":
        print("science fiction")
        user_choice = input("Provide a starting point or idea for the story")
        pronpt = f'''The cryopod hissed open, spewing a plume of frosty vapor that clung to {user_name} lashes. Disoriented, she stumbled out, gravity a forgotten friend after a century adrift. The stars, once pinpricks on velvet, were now an overwhelming ocean, her alien destination a colossal leviathan looming in the cosmic abyss. (Imagine {user_name} emerging from a cryopod aboard a spaceship, staring out at the vast expanse of stars and a nearby, enormous space station.){user_choice}'''
        res =generator(prompt,max_length=150,do_sample=True,temperature=0.9)
        print(res[0]['generated_text'])













if __name__ == "__main__":
    main()
```
